3. Considerando uma amostra de 45 números que representam o index do dataframe, índices= ([909, 751, 402, 400, 726, 39, 184, 269, 255, 769, 209, 715, 677, 381, 793, 697, 89, 280, 232, 756, 358, 36, 439, 768, 967, 699, 473, 222, 89, 639, 883, 558, 757, 84, 907, 895, 217, 224, 311, 348, 146, 505, 273, 957, 362]). Considerando essa amostra é possível dizer que a idade média das pessoas com problemas cardíacos é maior que 50 anos? Nível de significância igual a 5%.

```python
df_pacientes = pd.read_csv('pacientes.csv')
```

```python
df_pacientes.head(5)
```

```python
indices = [909, 751, 402, 400, 726, 39, 184, 269, 255, 769,
           209, 715, 677, 381, 793, 697, 89, 280, 232, 756,
           358, 36, 439, 768, 967, 699, 473, 222, 89, 639,
           883, 558, 757, 84, 907, 895, 217, 224, 311, 348,
           146, 505, 273, 957, 362]

amostra = df_pacientes.loc[indices]

media_amostra = amostra['Idade'].mean()
desvio_padrao_amostra = amostra['Idade'].std(ddof=1)  # ddof=1 para amostra

n = len(amostra)

media_populacao = 50
t_statistic = (media_amostra - media_populacao) / (desvio_padrao_amostra / np.sqrt(n))
p_value = 1 - stats.t.cdf(t_statistic, df=n-1)

print(f'Média da amostra: {media_amostra:.2f}')
print(f'Desvio padrão da amostra: {desvio_padrao_amostra:.2f}')
print(f'Estatística t: {t_statistic:.2f}')
print(f'Valor p: {p_value:.2f}')

if p_value < 0.05:
    print("Rejeita a hipótese nula: A idade média é maior que 50 anos.")
else:
    print("Aceita a hipótese nula: A idade média não é maior que 50 anos.")
```

6. Existe uma diferença significativa na pressão arterial média entre diferentes grupos étnicos nesta população? (Teste ANOVA, alpha é 5%) a. Hipótese Nula (H0 ): A pressão arterial média é a mesma em todos os grupos étnicos. b. Hipótese Alternativa (H1 ): Há uma diferença significativa na pressão arterial média entre pelo menos dois grupos étnicos nesta população.

```python
df_pacientes.head(2)
```

```python
grupos_etnicos = df_pacientes['Nome_Etnia'].unique()
print(f'Grupos étnicos encontrados: {grupos_etnicos}')

# Realizar teste ANOVA
f_statistic, p_value = stats.f_oneway(
    *[grupo['Pressao_Arterial'].values for nome, grupo in df_pacientes.groupby('Nome_Etnia')]
)

print(f'Estatística F: {f_statistic:.2f}')
print(f'Valor p: {p_value:.2f}')

if p_value < 0.05:
    print("Rejeita a hipótese nula: Há uma diferença significativa na pressão arterial média entre os grupos étnicos.")
else:
    print("Não rejeita a hipótese nula: A pressão arterial média é a mesma em todos os grupos étnicos.")
```

7. Lúcia é uma pesquisadora e tem o objetivo de entender a relação de gênero neste grupo de pacientes. Acredita-se que há uma relação entre o sexo e condições de saúde adicionais. (Teste qui-quadrado)

```python
df_pacientes.head(5)
```

```python
tabela_contingencia = pd.crosstab(df_pacientes['Genero'], df_pacientes['Nome_Estado_Saude'])

print("Tabela de Contingência:")
print(tabulate(tabela_contingencia, headers='keys', tablefmt='psql'))

chi2_statistic, p_value, dof, expected = stats.chi2_contingency(tabela_contingencia)

print(f'Estatística qui-quadrado: {chi2_statistic:.2f}')
print(f'Valor p: {p_value:.4f}')
print(f'Degrees of Freedom: {dof}')
print('Frequências esperadas:')
print(tabulate(expected, tablefmt='psql'))

if p_value < 0.05:
    print("Rejeita a hipótese nula: Há uma relação entre gênero e condições de saúde adicionais.")
else:
    print("Não rejeita a hipótese nula: Não há relação entre gênero e condições de saúde adicionais.")
```

8. Existe uma associação entre a idade dos pacientes e sua pressão arterial? a. Hipótese Nula (H0 ): se a pressão arterial é independente da idade b. Hipótese Alternativa (H1 ): a pressão arterial esta associada a idade

```python
df_pacientes.head(5)
```

```python
correlation_coefficient, p_value = pearsonr(df_pacientes['Idade'], df_pacientes['Pressao_Arterial'])

print(f'Coeficiente de Correlação de Pearson: {correlation_coefficient:.2f}')
print(f'Valor p: {p_value:.2f}')

alpha = 0.05
if p_value < alpha:
    print("Rejeita a hipótese nula: Existe uma associação entre a idade dos pacientes e pressão arterial.")
else:
    print("Não rejeita a hipótese nula: A pressão arterial é independente da idade dos pacientes.")
```

9. Qual é o intervalo de confiança para a média da pressão arterial entre os pacientes com condições de saúde adicionais? (nível de confiança 95%)

```python
df_pacientes.head(5)
```

```python
grupo_com_condicoes = df_pacientes[df_pacientes['Nome_Estado_Saude'] == 'Com condições de saúde adicionais']

media = grupo_com_condicoes['Pressao_Arterial'].mean()
desvio_padrao = grupo_com_condicoes['Pressao_Arterial'].std(ddof=1)
n = len(grupo_com_condicoes)


nivel_confianca = 0.95
z = stats.norm.ppf((1 + nivel_confianca) / 2)
margem_erro = z * (desvio_padrao / np.sqrt(n))
intervalo_conf = (media - margem_erro, media + margem_erro)

print(f'Média da pressão arterial: {media:.2f}')
print(f'Desvio padrão: {desvio_padrao:.2f}')
print(f'Tamanho da amostra: {n}')
print(f'Intervalo de Confiança (95%): ({intervalo_conf[0]:.2f}, {intervalo_conf[1]:.2f})')
```

10. A distribuição da pressão arterial na população segue uma distribuição normal? a. Hipótese Nula (H0 ): A distribuição da pressão arterial na população segue uma distribuição normal. b. Hipótese Alternativa (H1 ): A distribuição da pressão arterial na população não segue uma distribuição normal.

```python
df_pacientes.head(5)
```

```python
pressao_arterial = df_pacientes['Pressao_Arterial']

statistic, p_value = stats.shapiro(pressao_arterial)

print(f'Estatística do teste de Shapiro-Wilk: {statistic:.2f}')
print(f'Valor p: {p_value:.2f}')

alpha = 0.05  #
if p_value < alpha:
    print("Rejeita a hipótese nula: A distribuição da pressão arterial na população não segue uma distribuição normal.")
else:
    print("Não rejeita a hipótese nula: A distribuição da pressão arterial na população segue uma distribuição normal.")
```
